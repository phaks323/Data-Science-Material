## Natural Language Processing: Definition and steps
Automated Natural Language Processing (ANLP), also known as Natural Language Processing (NLP), is one of the most active research areas in data science today. It is a field at the intersection of Deep Learning and linguistics. Its goal is to extract semantic information from text content.

Natural Language Processing (NLP) has been used in many real-world applications, such as:

- Machine Translation;
- Spelling Check;
- Speech Recognition;
- Text Classification;
- Extract named entities from text;
- ...


The NLP pipeline generally consists of three main steps:

1. **Pre-processing:** a step to normalize the text and make it easier to use in the next stages of the algorithm;
2. **Represent text as a vector:** this step can be performed using Bag of Words (BoWs) or Term Frequency-Inverse Document Frequency (Tf-IdF) techniques. The vector representation (i.e. word embedding) can also be learned by Deep Learning algorithms.
**3. Classification:** a task of mapping out similar sentences.


This guide will cover the most common NLP tasks where specific tools are involved. All these tasks are mainly done with SpaCy, an open-source software library that provides pre-trained models for a variety of applications, including named entity recognition. [SpaCy](https://spacy.io/) is an NLP library that uses Python, which is the main alternative to  [NLTK (Natural Language Tool Kit)](https://www.nltk.org/), offering a lot of features and possibilities that make it very useful.

### 1. Tokenization

Tokenization attempts to convert text into a series of distinct tokens. The idea is that each token represents a word vector, and identifying words seems like a relatively simple task. But what do we do with an example like "I'm cold" in Italy? In the tokenization model, we need to split "I" as the first word.

SpaCy offers a tokenization function using the *nlp* method. This method is the entry point for all SpaCy functions. It is used to display text in a form that can be interpreted by the library.

```
def return_tokens(stc):
    # Tokenize the sentence
    doc = nlp(stc)
    # Return the text of each token
    return [x.text for x in doc]
```

````
def return_tokens(stc):
    # Tokenize the sentence
    doc = nlp(stc)
    # Return the text of each token
    return [x.text for x in doc]
````

### 3. Sentence tokenization 

Sentence tokenization can also be used to identify different sentences in a piece of text. A priori, this step may seem simple, as it is sufficient to cut each sentence into words when a dot appears.
Therefore, to achieve good tokenization, you need to understand its full context.


```
def return_token_stc(stc):
    # Tokenize the sentence
    doc = nlp(stc)
    # Retourner le texte de chaque phrase
    return [x.text for x in doc.sents]
```

Let's apply the tokenization scheme to the following sentence:


> Buongiorno. Parla inglese? Parli inglese?





````
def return_token_stc(stc):
    # Tokenize the sentence
    doc = nlp(stc)
    # Retourner le texte de chaque phrase
    return [x.text for x in doc.sents]
````

### 4. Stemming

Stemming consists of reducing a word to its root form. The purpose of stemming is to group the many variations of a word into a single word. For example, once stemming is applied to "Dogs" or "Dog", the result is the same word. This reduces the size of the vocabulary in the word bag model or Tf-IdF approach.


The most famous stemmer was the *Snowball Stemmer*. It is available in Italian.


```
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language='italian')

def return_stem(stc):
    doc = nlp(stc)
    return [stemmer.stem(x.text) for x in doc]
```



````
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language='italian')

def return_stem(stc):
    doc = nlp(stc)
    return [stemmer.stem(x.text) for x in doc]
````

````
def return_NER(stc):
    # Tokenize the sentence
    doc = nlp(stc)
    # Return the text and label for each entity
    return [(x.text, x.label_) for x in doc.ents]
````

### 8. Sentence Similarity


To judge the degree of similarity between two sentences, we will use a very simple method:

- Determine the average embedding of a given text by embedding all words in the sentence.
- Calculate the distance between two sentences by simple Euclidean distance.

With SpaC

```
# Ce texte est au format code
```

y, it is so easy!


```
def return_mean_embedding(stc):
    # Tokenize the sentence
    doc = nlp(stc)
    # Returning the average of the vectors for each sentence
    return np.mean([(x.vector) for x in doc], axis=0)
```



````
def return_mean_embedding(stc):
    # Tokenize the sentence
    doc = nlp(stc)
    # Returning the average of the vectors for each sentence
    return np.mean([(x.vector) for x in doc], axis=0)
````

### 9. Transformers

Transformers are particularly well suited for these purposes:

- Determine if a sentence is the continuation of another sentence.
